- åœ¨ä»¥ä¸‹ç¯„ä¾‹ä¸­ï¼Œtri_recursion() æ˜¯æˆ‘å€‘å®šç¾©çš„å‡½æ•¸ï¼Œç”¨æ–¼å‘¼å«è‡ªå·±ï¼ˆrecurseï¼‰ã€‚  æˆ‘å€‘ä½¿ç”¨ k è®Šæ•¸ä½œç‚ºæ•¸æ“šï¼Œæ¯æ¬¡éè¿´æ™‚å®ƒéƒ½æœƒéæ¸› (-1)ã€‚  ç•¶æ¢ä»¶ä¸å¤§æ–¼0æ™‚ï¼ˆå³ç‚º0æ™‚ï¼‰éè¿´çµæŸ


```python
def tri_recursion(k):
  if(k > 0):
    result = k + tri_recursion(k - 1)
    print(result)
  else:
    result = 0
  return result
```
